```python
#k-means clustering 10.5.1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#initialize random samples same as the lab
np.random.seed(5)
Xsamples = np.random.randn(50,2)
Xsamples[0:25, 0] = Xsamples[0:25, 0] + 3
Xsamples[0:25, 1] = Xsamples[0:25, 1] - 4
```

```python
#Use k-means from sklearn on the samples, with 2 clusters
kmeans = KMeans(n_clusters=2).fit(Xsamples)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
plt.scatter(Xsamples[:,0],Xsamples[:,1], c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],label='ClusterCenter')
```

```python
#Use k-means from sklearn on the samples, with 3 clusters
kmeans2 = KMeans(n_clusters=3).fit(Xsamples)
plt.scatter(Xsamples[:,0],Xsamples[:,1], c = kmeans2.labels_)
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],label='ClusterCenter')
```

```python
#Use k-means from sklearn on the samples, with 3 clusters
kmeans2 = KMeans(n_clusters=3, n_init = 1,max_iter = 40).fit(Xsamples)
plt.scatter(Xsamples[:,0],Xsamples[:,1], c = kmeans2.labels_)
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],label='ClusterCenter')
```

```python
#Use k-means from sklearn on the samples, with 3 clusters
kmeans2 = KMeans(n_clusters=3, n_init = 20, max_iter = 40).fit(Xsamples)
plt.scatter(Xsamples[:,0],Xsamples[:,1], c = kmeans2.labels_)
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],label='ClusterCenter')
```

```python
#Hierarchical clustering
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage
Complete_linkage = linkage(Xsamples,"complete")

plt.title('complete linkage')
plt.xlabel('index of samples')
plt.ylabel('distance')
dendrogram(Complete_linkage)
```

```python
Average_linkage = linkage(Xsamples,"average")
plt.title('Average linkage')
plt.xlabel('index of samples')
plt.ylabel('distance')
dendrogram(Average_linkage)
```

```python
Single_linkage = linkage(Xsamples,"single")
plt.title('single linkage')
plt.xlabel('index of samples')
plt.ylabel('distance')
dendrogram(Single_linkage)
```

```python
from scipy.cluster.hierarchy import cut_tree
print(cut_tree(Complete_linkage, n_clusters = 2).T)
print(cut_tree(Average_linkage, n_clusters = 2).T)
print(cut_tree(Single_linkage, n_clusters = 2).T)
```

```python
#try with 3 dimensions for complete linkage
np.random.seed(5)
Xsamples3d = np.random.randn(50,3)
CompleteCorrelationLinkage = linkage(Xsamples3d, 'complete', 'correlation')
plt.title('complete linkage correlation based')
plt.xlabel('index of samples')
plt.ylabel('distance')
dendrogram(CompleteCorrelationLinkage)
```

```python
CompleteCorrelationLinkage = linkage(Xsamples3d, 'complete')
plt.title('complete linkage euclid based')
plt.xlabel('index of samples')
plt.ylabel('distance')
dendrogram(CompleteCorrelationLinkage)
```
